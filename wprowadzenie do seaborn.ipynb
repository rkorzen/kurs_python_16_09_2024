# Wprowadzenie do Seaborn

## 1. Czym jest Seaborn?

Seaborn to biblioteka Pythona oparta na Matplotlib, służąca do tworzenia atrakcyjnych i informatywnych wizualizacji danych statystycznych. Ułatwia tworzenie skomplikowanych wykresów, oferując proste interfejsy do pracy z danymi. Seaborn integruje się również z bibliotekami takimi jak Pandas, co umożliwia bezpośrednie wizualizowanie danych z **DataFrame**.

## 4. Rodzaje wykresów w Seaborn

### 4.1. Wykresy rozrzutu (Scatter Plots)

#### `scatterplot`

Używany do wizualizacji relacji między dwiema zmiennymi ciągłymi.



````python
# Przykładowe dane
tips = sns.load_dataset('tips')

# Wykres rozrzutu
sns.scatterplot(data=tips, x='total_bill', y='tip')

plt.title('Wykres rozrzutu - napiwek vs. rachunek')
plt.xlabel('Rachunek ($)')
plt.ylabel('Napiwek ($)')
plt.show()
````

### 4.2. Wykresy liniowe

#### `lineplot`

Przydatny do wizualizacji trendów w danych ciągłych.

````python
sns.lineplot(data=tips, x='size', y='total_bill')

plt.title('Średni rachunek w zależności od liczby osób')
plt.xlabel('Liczba osób')
plt.ylabel('Średni rachunek ($)')
plt.show()
````

### 4.3. Wykresy pudełkowe (Box Plots)

#### `boxplot`

Używane do przedstawienia rozkładu danych i identyfikacji wartości odstających.

````python
sns.boxplot(data=tips, x='day', y='total_bill')

plt.title('Rozkład rachunków w poszczególne dni')
plt.xlabel('Dzień tygodnia')
plt.ylabel('Rachunek ($)')
plt.show()
````

### 4.4. Histogramy i wykresy gęstości

#### `histplot`

Tworzy histogram danych.

````python
sns.histplot(data=tips, x='total_bill', bins=30)

plt.title('Histogram rachunków')
plt.xlabel('Rachunek ($)')
plt.ylabel('Liczba obserwacji')
plt.show()
````

#### `kdeplot`

Wykres estymaty gęstości jądra (Kernel Density Estimate).

````python
sns.kdeplot(data=tips['total_bill'], shade=True)

plt.title('Wykres gęstości rachunków')
plt.xlabel('Rachunek ($)')
plt.ylabel('Gęstość')
plt.show()
````

### 4.5. Wykresy słupkowe (Bar Plots)

#### `barplot`

Przedstawia średnie wartości zmiennej ciągłej w kategoriach.

````python
sns.barplot(data=tips, x='day', y='total_bill', ci='sd')

plt.title('Średni rachunek w poszczególne dni')
plt.xlabel('Dzień tygodnia')
plt.ylabel('Średni rachunek ($)')
plt.show()
````

### 4.6. Wykresy punktowe (Point Plots)

Pokazują zmiany średnich wartości z uwzględnieniem innej zmiennej.

````python
sns.pointplot(data=tips, x='day', y='total_bill', hue='sex')

plt.title('Średni rachunek w poszczególne dni z podziałem na płeć')
plt.xlabel('Dzień tygodnia')
plt.ylabel('Średni rachunek ($)')
plt.legend(title='Płeć')
plt.show()
````

### 4.7. Wykresy relacji par (Pair Plots)

#### `pairplot`

Tworzy siatkę wykresów dla wszystkich par zmiennych.

````python
sns.pairplot(tips)

plt.show()
````

### 4.8. Mapy cieplne (Heatmaps)

Używane do wizualizacji macierzy korelacji lub innych danych 2D.

````python
corr = tips.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title('Mapa cieplna korelacji')
plt.show()
````

### 5.2. Używanie `hue` do grupowania danych

Wiele funkcji Seaborn posiada parametr `hue`, który pozwala na rozróżnienie danych według kategorii.

````python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')

plt.title('Napiwek vs. Rachunek z podziałem na płeć')
plt.show()
````

## 6. Zaawansowane wykresy w Seaborn

### 6.1. Wykresy regresji

#### `lmplot`

Tworzy wykresy z linią regresji.

````python
sns.lmplot(data=tips, x='total_bill', y='tip', hue='sex')

plt.title('Regresja napiwku względem rachunku')
plt.show()
````

````python
### 6.2. Wykresy dystrybuant (Distribution Plots)

#### `displot`

Tworzy histogram i wykres gęstości jednocześnie.
````

````python
sns.displot(tips['total_bill'], kde=True)

plt.title('Dystrybucja rachunków')
plt.xlabel('Rachunek ($)')
plt.ylabel('Gęstość')
plt.show()
````

````python
### 6.3. Wykresy skrzypcowe (Violin Plots)

Przedstawiają rozkład danych z uwzględnieniem gęstości.

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)

plt.title('Rozkład rachunków w poszczególne dni z podziałem na płeć')
plt.show()
```
````

### 6.4. Facetowanie wykresów

#### `FacetGrid`

Pozwala na tworzenie siatek wykresów w oparciu o kategorie.

````python
g = sns.FacetGrid(tips, col='day', hue='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
plt.show()
````

## 8. Przykłady praktyczne

### 8.1. Analiza danych o diamentach

````python
diamonds = sns.load_dataset('diamonds')

# Wykres relacji ceny do masy kamienia z uwzględnieniem czystości
sns.scatterplot(data=diamonds, x='carat', y='price', hue='clarity', alpha=0.5)

plt.title('Cena vs. Masa diamentu z uwzględnieniem czystości')
plt.show()
````

### 8.2. Analiza danych medycznych

````python
fmri = sns.load_dataset('fmri')

sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')

plt.title('Sygnał fMRI w czasie')
plt.show()
````
