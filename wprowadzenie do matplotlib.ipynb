# Wprowadzenie do Matplotlib

## 1. Czym jest Matplotlib?

Matplotlib to biblioteka Pythona służąca do tworzenia wysokiej jakości wykresów 2D (i prostych 3D). Umożliwia wizualizację danych w różnych formatach i jest szeroko stosowana w nauce, inżynierii oraz analizie danych. Matplotlib jest podstawą dla bibliotek takich jak Seaborn czy Pandas (plotting).

## 3. Podstawy korzystania z Matplotlib

### 3.1. Importowanie biblioteki

Standardowo Matplotlib importuje się jako:

````python
import matplotlib.pyplot as plt
````

````python
### 3.2. Tworzenie prostego wykresu liniowego
````

````python
import matplotlib.pyplot as plt

# Dane
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Tworzenie wykresu
plt.plot(x, y)

# Wyświetlanie wykresu
plt.show()
````

## 4. Rodzaje wykresów

### 4.1. Wykres liniowy (`plot`)

Używany do przedstawiania danych ciągłych.

````python
plt.plot(x, y, label='Dane')
````

### 4.2. Wykres punktowy (`scatter`)

Przydatny do przedstawiania zależności między dwiema zmiennymi.

````python
plt.scatter(x, y, label='Punkty', color='red')
````

### 4.3. Histogram (`hist`)

Używany do przedstawiania rozkładu danych.

````python
import numpy as np

dane = np.random.randn(1000)
plt.hist(dane, bins=30, label='Histogram')
````

### 4.4. Wykres słupkowy (`bar`)

Przydatny do przedstawiania danych kategorycznych.

````python
kategorie = ['A', 'B', 'C', 'D']
wartości = [5, 7, 3, 8]
plt.bar(kategorie, wartości, label='Słupki')
````

### 4.5. Wykres kołowy (`pie`)

Używany do przedstawiania proporcji.

````python
udziały = [50, 25, 15, 10]
etykiety = ['A', 'B', 'C', 'D']
plt.pie(udziały, labels=etykiety, autopct='%1.1f%%')
````

### 4.6. Wykres pudełkowy (`boxplot`)

Przedstawia rozkład danych na podstawie kwartylów.

````python
dane = [np.random.randn(100) for _ in range(4)]
plt.boxplot(dane, labels=['Grupa 1', 'Grupa 2', 'Grupa 3', 'Grupa 4'])
````

````python
### 4.7. Wykres 3D
````

````python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z)
````

## 6. Przykłady

### 6.1. Wielokrotne serie danych na jednym wykresie

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 9]

plt.plot(x, y1, label='Seria 1')
plt.plot(x, y2, label='Seria 2')

plt.title('Wykres z wieloma seriami')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.legend()
plt.grid(True)
plt.show()
````

### 6.2. Podwykresy (Subplots)

````python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Wykres 1
x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sinus')

# Wykres 2
axs[0, 1].scatter(np.random.rand(50), np.random.rand(50), color='red')
axs[0, 1].set_title('Punkty losowe')

# Wykres 3
axs[1, 0].bar(kategorie, wartości, color='green')
axs[1, 0].set_title('Słupki')

# Wykres 4
dane = np.random.randn(1000)
axs[1, 1].hist(dane, bins=30)
axs[1, 1].set_title('Histogram')

plt.tight_layout()
plt.show()
````

### 6.3. Wykres punktowy z kolorami i rozmiarami

````python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
kolory = np.random.rand(50)
rozmiary = 1000 * np.random.rand(50)

plt.scatter(x, y, c=kolory, s=rozmiary, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.show()
````

````python
### 6.4. Wykres funkcji matematycznej
````

````python
import numpy as np

x = np.linspace(-10, 10, 400)
y = np.sin(x) / x

plt.plot(x, y)
plt.title('Wykres funkcji sin(x)/x')
plt.xlabel('x')
plt.ylabel('sin(x)/x')
plt.grid(True)
plt.show()
````

## 8. Praca z datami

````python
import numpy as np
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import datetime

daty = [datetime.datetime(2021, 1, i+1) for i in range(10)]
wartości = np.random.rand(10)

plt.plot(daty, wartości)
plt.title('Wykres z datami')
plt.xlabel('Data')
plt.ylabel('Wartość')

# Formatowanie osi X
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()

plt.show()
````
